### ðŸ“–âœ¨ New Terminology:
- __default parameters__: setting a parameter to have a default value so you don't always have to pass it into the function.

In the example below, we set `tip_percent=0.20` in the function definition, so `tip_percent` has a default value of `0.20`, which means it will be 0.20 unless otherwise specified.

```python
def calculate_tip(bill, tip_percent=0.20):
    """
    will return the amount you should tip given the total of the bill and the tip_percent
    
    Parameters:
        bill, (float), the total amount the bill is before the tip
        tip_percent, (float), default value of 0.2, the amount you want to tip your server.
    
    Return:
        the amount you should tip given the parameters
    """
    return bill*tip_percent
```

When we can call `calculate_tip(100.0)` below without specifying the `tip_percent` argument, we don't get an error-- it automatically assumes `tip_percent=0.20`.

```python
print("On a $100 bill, you should tip, $", calculate_tip(100.0), sep='')
```

But, if you'd like to use something other than the __default parameter__ value (in this case, `0.20`), you can override the default parameter by passing in a different value as an argument:

```python
print("If you get terrible service, you should tip $", calculate_tip(100, 0.10), sep='')
```

### Multiple Parameters Review
So far, when dealing with multiple arguments/parameters, we've learned that the order is important!
- `parameter 1` will take on the value of `argument 1`, `parameter 2` will take on the value of `argument 2`, etc.

```python
def calculate_pay(wage, hours, overtime_threshold):
    if hours < overtime_threshold:
        return hours*wage
    else:
        return 40*wage + (hours-overtime_threshold) * wage * 1.5
    
print(calculate_pay(14.5, 45, 40))
```

### ðŸ“–âœ¨ New Terminology:
But... this is not always the case.
- __keyword arguments__: when calling a function with multiple parameters, you can specify the parameter by the keyword so that you don't have to worry about the order of the arguments.

```python
def calculate_pay(wage, hours, overtime_threshold):
    if hours < overtime_threshold:
        return hours*wage
    else:
        return 40*wage + (hours-overtime_threshold) * wage * 1.5
    
print(calculate_pay(14.5, 45, 40))
print(calculate_pay(hours=45, wage=14.5, overtime_threshold=40))
```

# Count-Controlled Loops

__count-controlled loop__: a loop that ierates a specific number of times. 
- So far, we've accomplished this using a combination of a `counter`, `while` loop, and the `len()` function.


```python
days_in_week = ['Sunday','Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
counter = 0
while counter < len(days_in_week):
    print(days_in_week[counter])
    counter +=1
```

# The For Loop:
## ðŸ”âœ¨ New Keywords Unlocked: `for` and `in`
Use a `for` statement to easily write count-controlled loops.
- designed to work with a sequence of data items.
    - iterates once for each item in the sequence.

## `for` loop syntax:
- keyword `for`
- __target variable__: you get to make this one up, but it's best if it hasn't been used in your current program.
- keyword `in`
- a sequence
- colon `:`

Example:

```python
days_in_week = ['Sunday','Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

for day in days_in_week: #day is the target variable
    print(day)
```

```python
for num in [1,2,3,4,5]:
    print(num)
```

# The `range()` function

The `range()` function simlifies the process of writing a for loop.

- `range()` returns an iterable object
    - *iterable* means that it contains a sequence of values that can be iterated (looped) over.

```python
range(5)
```

```python
for num in range(5):
    print(num)
```

```python
for x in range(5):
    print("Hello World!")
```

## Passing `range()` Multiple Arguments

By default, the range function produces a sequence of numbers that increases by 1.

If you pass three arguments to range:
- the first number is the starting value
- the second number is the ending limit
- the third number is the step value (increment)

```python
for x in range(1,10,2):
    print(x)
```

## Two Main Formulas for For Loops:

There are two common strategies for using for loops with lists. 

The first is called `list_comprehension`, and is used when you want to have access only to the values that are in the list, but don't necessarily care about the indices of the list. 

It looks something like this:

```python
temperatures = [30, 20, 2, -5, -15, -8, -1, 0, 5, 35]

for temp in temperatures:
    if temp < 0:
        print(temp)
```

What if we wanted to print out which days had negative temperatures? Our output might look like:

`Day 3 had a negative temperature`

`Day 4 had a negative temperature`

`Day 5 had a negative temperature`

`Day 6 had a negative temperature`

Any ideas how we might accomplish this?

Possible solutions:
- use a counter
- use the `range()` function

```python
# what will this code do?
for index in range(len(temperatures)):
    print(index)
```

```python
temperatures = [30, 20, 2, -5, -15, -8, -1, 0, 5, 35]

for index in range(len(temperatures)):
    if temperatures[index] < 0:
        print("Day", index, " had a negative temperature")
```
