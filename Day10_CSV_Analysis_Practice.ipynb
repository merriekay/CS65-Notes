```python
import csv

with open("nationalparks.csv") as npfile:
    parks = csv.reader(npfile)
    parks = list(parks)
    
state = input("Enter a state: ")

park_counter = 1
parks_in_state = 0

while park_counter < len(parks):
    
    if parks[park_counter][1] == state:
        parks_in_state += 1
    
    
    park_counter += 1
    
print("There are",parks_in_state,"national parks in",state)
```

## Group Exercise #1
Consider the following program from above that created a 2D list of national park info from a csv file and then counted how many parks there are in a given state.

```python
import csv

with open("nationalparks.csv") as npfile:
    parks = csv.reader(npfile)
    parks = list(parks)
    
state = input("Enter a state: ")

park_counter = 1
parks_in_state = 0

while park_counter < len(parks):
    
    if parks[park_counter][1] == state:
        parks_in_state += 1
    
    
    park_counter += 1
    
print("There are",parks_in_state,"national parks in",state)
```
